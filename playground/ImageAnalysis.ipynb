```python
from skimage.feature import ORB, match_descriptors

detector_extractor1 = ORB(n_keypoints=5)
detector_extractor2 = ORB(n_keypoints=5)
detector_extractor3 = ORB(n_keypoints=5)
detector_extractor4 = ORB(n_keypoints=5)

detector_extractor1.detect_and_extract(color.rgb2gray(img1))
detector_extractor2.detect_and_extract(color.rgb2gray(img2))
detector_extractor3.detect_and_extract(color.rgb2gray(img3))
detector_extractor4.detect_and_extract(color.rgb2gray(img4))

matches = match_descriptors(detector_extractor1.descriptors, detector_extractor2.descriptors)

print detector_extractor1.keypoints.ravel()
print detector_extractor2.keypoints
print detector_extractor3.keypoints
# print matches
# print detector_extractor1.keypoints[matches[:, 0]]
# print detector_extractor2.keypoints[matches[:, 1]]
```

```python
print "Cosine img 2, img 3", cosine_similarity(color.rgb2gray(detector_extractor4.keypoints).reshape(1, -1), color.rgb2gray(detector_extractor3.keypoints).reshape(1, -1))
print "Euclidean img 2, img 3", euclidean_distances(color.rgb2gray(detector_extractor4.keypoints).reshape(1, -1), color.rgb2gray(detector_extractor3.keypoints).reshape(1, -1))
```
